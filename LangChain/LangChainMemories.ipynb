```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
import os

GEMINI_API_KEY = "AIzaSyAcgfiBbaZ56qrZDpqfIeBvKS9x4znu-1Y"  # Replace with your actual Gemini API key
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-pro",
    google_api_key=GEMINI_API_KEY,
    temperature=0
)
```

**Buffer Memory**

```python
prompt_template = PromptTemplate(
    input_variables=["history", "input"],
    template="""
    You are a helpful AI assistant. Use the following conversation history to answer the user's question.

    Conversation History:
    {history}

    User: {input}
    AI:
    """
)

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory,
    prompt=prompt_template,
    verbose=True
)
```

```python
print("Start chatting with the AI! Type 'exit' to end the conversation.")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break

    # Get AI response
    response = conversation.predict(input=user_input)
    print(f"AI: {response}")
```

```python

# Display memory buffer (optional)
print("\nMemory Buffer:")
print(memory.buffer)
print("-" * 50)
```

**ConversationBufferWindowMemory**

the parameter k refers to the size of the sliding window for conversation history. It determines how many of the most recent interactions (user inputs and AI responses) are kept in memory.

```python
from langchain.memory import ConversationBufferWindowMemory  # Sliding window memory

memory = ConversationBufferWindowMemory(k=2)  # Adjust `k` for the desired window size

# Create a conversation chain with buffer window memory
conversation = ConversationChain(
    llm=llm,
    memory=memory,
    prompt=prompt_template,
    verbose=True  # Set to True to see the internal workings
)
# Example conversation
print("Start chatting with the AI! Type 'exit' to end the conversation.")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break

    # Get AI response
    response = conversation.predict(input=user_input)
    print(f"AI: {response}")

```

```python
# Display memory buffer
print("\nMemory Buffer:")
print(memory.buffer)
```

**ConversationTokenBufferMemory**

This memory type retains conversation history up to a specified number of tokens, ensuring that the memory doesn't exceed the token limit of the model

```python
from langchain.memory import ConversationTokenBufferMemory

memory = ConversationTokenBufferMemory(
    llm=llm,
    max_token_limit=50
)

conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)

questions = [
    "Hi",
    "What's the weather like?",
    "Tell me a joke",
    "Do you remember my name?"
]

for question in questions:
    response = conversation.predict(input=question)
    print(f"AI: {response}")

    # Display memory buffer
    print("\nMemory Buffer:")
    print(memory.buffer)
    print("-" * 50)
```

**ConversationSummaryMemory**

this memory type is designed to maintain a summary of the conversation history rather than storing the entire history verbatim. This is particularly useful for long conversations where retaining the full history would be impractical due to token limits or memory constraints.

```python
from langchain.memory import ConversationSummaryMemory

memory = ConversationSummaryMemory(llm=llm)

conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)

# Simulate a conversation
questions = [
    "Hi",
    "What's the weather like?",
    "Tell me a joke",
    "Do you remember my name?"
]

for question in questions:
    response = conversation.predict(input=question)
    print(f"AI: {response}")

    # Display conversation summary
    print("\nConversation Summary:")
    print(memory.buffer)
    print("-" * 50)
```
